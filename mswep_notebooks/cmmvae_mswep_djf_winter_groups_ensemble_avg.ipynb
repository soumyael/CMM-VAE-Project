```python
# Ensemble Averaging of All 5 Folds
import numpy as np
# Load fold data
folds = [
    np.load('fold1.npy'),
    np.load('fold2.npy'),
    np.load('fold3.npy'),
    np.load('fold4.npy'),
    np.load('fold5.npy')
]
# Average the folds
ensemble_avg = np.mean(folds, axis=0)

# Save the ensemble average
np.save('ensemble_average.npy', ensemble_avg)
```
